```python
class QueADT :
  def __init__(self, _max = 10):
    self.dQueList=[]
    self.front = 0
    self.rear = 0
    self._max = _max+1
  
  def Deque(self):
    self.dQueList = [None]*(self._max)
  
  def isEmpty(self):
    if self.front == self.rear:
      print("덱이 비어있습니다.")
      return True
    else:
      print("덱이 비어있지 않습니다.")
      return False

  def addFront(self, x):
    if self.front != (self.rear+1)%self._max:
      self.dQueList[self.front] = x
      self.front = self.front-1
      if self.front < 0:
        self.front = self._max-1
      print("{}을/를 덱의 맨 앞에 추가하였습니다.".format(x))
    else:
      print("덱이 포화상태입니다.")

  def deleteFront(self):
    if self.front != self.rear:
      self.front = (self.front+1)%self._max
      _item = self.dQueList[self.front]
      print("덱의 맨 앞에서 {}을/를 꺼냈습니다.".format(_item))
    else:
      print("덱이 비어있습니다.")
  
  def getFront(self):
    if self.front != self.rear:
      temp = (self.front+1)%self._max
      _item = self.dQueList[temp]
      print("덱의 맨 앞에 있는 항목은 {}입니다.".format(_item))
    else:
      print("덱이 비어있습니다.")
  
  def addRear(self, x):
    if self.front !=  (self.rear+1)%self._max:
      self.rear = (self.rear+1) % self._max
      self.dQueList[self.rear] = x
      print("{}을/를 덱의 맨 뒤에 추가하였습니다.".format(x))
    else:
      print("덱이 포화상태입니다.")
  
  def deleteRear(self):
    if self.front != self.rear:
      _item = self.dQueList[self.rear]
      self.rear = self.rear-1
      if self.rear < 0:
        self.rear = self._max-1
      print("덱의 맨 뒤에서 {}을/를 꺼냈습니다.".format(_item))
    else:
      print("덱가 비어있습니다.")
  
  def getRear(self):
    if self.front != self.rear:
      _item = self.dQueList[self.rear]
      print("덱의 맨 뒤에 있는 항목은 {}입니다.".format(_item))
    else:
      print("덱이 비어있습니다.")

  def isFull(self):
    if self.front == (self.rear+1)%self._max:
      print("덱이 포화상태입니다.")
      return True
    else:
      print("덱이 가득 차있지 않습니다.")
      return False
  
  def size(self):
    count = (self.rear - self.front + self._max) % self._max
    if count != 0:
      print("현재 덱의 항목은 총 {}개입니다.".format(count))
    else:
      print("덱이 비어있습니다.")
  
  def clear(self):
    self.front = self.rear
    print("덱을 비웠습니다.")
  
  def check(self):
    templist = []
    start = (self.front+1)%self._max
    stop = (self.rear+1)%self._max

    if stop < start:
      for i in range(start, self._max):
        templist.append(self.dQueList[i])
      for i in range(stop):
        templist.append(self.dQueList[i])
    else:
      for i in range(stop):
        templist.append(self.dQueList[i])
    print(templist)
```
