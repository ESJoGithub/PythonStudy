```python
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow

_file = files.upload()
```

```python
img_orgn = cv2.imread("flower.jpg")
height, width, channel = img_orgn.shape
img = cv2.resize(img_orgn, (height//2, width//2))
cv2_imshow(img)
```

```python
height, width, channel = img.shape
```

```python
def img_resize(_img, size):
  height, width, channel = _img.shape
  c_height = int(np.ceil(height*size))
  c_width = int(np.ceil(width*size))
  c_img = np.zeros(shape=(c_height, c_width, channel), dtype="uint8")

  index = []
  _size = 1/size

  for y in range(0, c_height):
    for x in range(0, c_width):
      c_img[y][x][0] = _img[int(y*_size)][int(x*_size)][0]
      c_img[y][x][1] = _img[int(y*_size)][int(x*_size)][1]
      c_img[y][x][2] = _img[int(y*_size)][int(x*_size)][2]

      index.append((y,x))
    
  cv2_imshow(c_img)   

```

```python
img_resize(img, 0.5)
```

```python
img_resize(img, 2.3)
```
