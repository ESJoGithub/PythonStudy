## **Scikit-learn 활용 예제 실습**

### Scikit-learn을 활용한 iris 분류 실습

```python
from sklearn.datasets import load_iris               # 사이킷런 내장 데이터set 중 iris 데이터 set 로드 모듈
from sklearn.tree import DecisionTreeClassifier      # 의사결정트리 알고리즘 모듈
from sklearn.neighbors import KNeighborsClassifier   # 최근접 이웃 알고리즘(K-NN) 모듈
from sklearn.model_selection import train_test_split # 데이터set을 Train/Test set으로 분리 모듈
from sklearn.metrics import accuracy_score           # 정확도 판별 모듈

import pandas as pd
import seaborn as sns
```

```python
iris = load_iris()                # iris 객체 생성
iris_data = iris.data             # iris 데이터(각 붓꽃의 특징)를 iris_data에 대입
iris_label = iris.target          # iris 분류를 iris_label에 대입
iris_df = pd.DataFrame(data = iris_data, columns = iris.feature_names) 
iris_df['label'] = iris.target

sns.pairplot(iris_df, 
             x_vars = ["sepal length (cm)"], 
             y_vars = ["sepal width (cm)"], 
             hue = "label", height = 5)
```

```python
"""iris_data(feature), iris_label(label) 기준으로 trins: test 비율을 8:2로 나눔 
 x_train(학습 데이터 feature) y_train (학습 데이터 label) 
 x_test(테스트 데이터 feature), y_test(테스트 데이터 label)"""

x_train, x_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size = 0.2)

dt_clf = DecisionTreeClassifier() # 의사결정트리 객체 생성
knn_clf = KNeighborsClassifier()  # K-NN 객체 생성

# 학습 진행
dt_clf.fit(x_train, y_train)
knn_clf.fit(x_train, y_train)

# 테스트 데이터 입력하여 결과 예측
y1_pred = dt_clf.predict(x_test)  
y2_pred = knn_clf.predict(x_test) 
```

```python
print("DecisionTree 예측정확도: {0: .4f}".format(accuracy_score(y_test, y1_pred)))
print("K-NN 예측정확도: {0: .4f}".format(accuracy_score(y_test, y2_pred)))
```
