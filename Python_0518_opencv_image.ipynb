### OpenCv

함수 사용을 최소화하여 이미지 리사이즈 구현하기

```python
import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow

file1 = files.upload()
```

```python
img1 = cv2.imread("flower.jpg")
height, width, channel = img1.shape
print(img1.shape)
img1[20:30][:][:] = 0
cv2_imshow(img1)
```

```python
height, width, channel = img1.shape
copy_img = np.zeros(shape=(int(np.ceil(height/2)), int(np.ceil(width/2)), channel), dtype=np.uint8)

for y in range(0, height, 2):
   for x in range(0, width, 2):
    copy_img[y//2][x//2][0] = img1[y][x][0]
    copy_img[y//2][x//2][1] = img1[y][x][1]
    copy_img[y//2][x//2][2] = img1[y][x][2]

cv2_imshow(copy_img)
```

```python
height, width, channel = img1.shape
c_height = height*2
c_width = width*2
copy_img = np.zeros(shape=(c_height, c_width, channel), dtype=np.uint8)

for y in range(c_height):
  for x in range(c_width):
    copy_img[y][x][0] = img1[y//2][x//2][0]
    copy_img[y][x][1] = img1[y//2][x//2][1]
    copy_img[y][x][2] = img1[y//2][x//2][2]

cv2_imshow(copy_img)
```
